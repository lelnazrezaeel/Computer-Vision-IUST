```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt

##################### your code ################################
img = cv2.imread("royan.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


################################################################
# Finding contours
## result = result otso|adaptive algorithm
ret,result1 = cv2.threshold(gray, 127, 255, cv2.THRESH_OTSU)
print("Otsu")
print("-------------------------------------------------")
print(pytesseract.image_to_string( result1 , lang='fas'))
print("-------------------------------------------------")
result2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 41, 5)
print("Adaptive theresholding")
print("-------------------------------------------------")
print(pytesseract.image_to_string( result2 , lang='fas'))
print("-------------------------------------------------")

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(14, 14),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(gray, cmap='gray')
ax1.set_title('Source')
ax2.imshow(result1, cmap='gray')
ax2.set_title('Otsu')
ax3.imshow(result2, cmap='gray')
ax3.set_title('Adaptive theresholding')

plt.tight_layout()
plt.show()
```

```python
image_path_in_colab='royan.jpg'
extractedInformation = pytesseract.image_to_string(image_path_in_colab, lang='fas')
print(extractedInformation)
```

```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
##################### your code ################################
img = cv2.imread("mypic.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,result = cv2.threshold(gray, 127, 255, cv2.THRESH_OTSU)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
dilation = cv2.dilate(result, kernel, iterations=1)

################################################################
# Finding contours
im2 = img.copy()
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(im2, (x, y), (x + w, y + h), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = im2[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))
```

```python
image_path_in_colab='mypic.png'
extractedInformation = pytesseract.image_to_string(image_path_in_colab, lang='fas')
print(extractedInformation)
```
